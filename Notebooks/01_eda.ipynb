```python
# Import libraries to be used 
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the dataset and preview the data as well as the metadata
df = pd.read_csv(r"C:\Users\Admin\OneDrive\Documents\Data Science Portfolio\Lead Scoring\Data\Lead Scoring.csv")
df.head()
df.info()
df.describe()
df.isnull().sum()
```

```python
# Analyse target variable
df['Converted'].value_counts(normalize = True).plot(kind = 'bar', color = ['skyblue', 'salmon'])
plt.title('Conversion Rate')
plt.xlabel('Converted')
plt.ylabel('Proportion')
plt.show()
```

```python
# Categorical Feature Exploration
cat_cols = df.select_dtypes(include = 'object').columns.tolist()
print(cat_cols)
```

```python
def plot_cats(cols):
    for col in cols:
        plt.figure(figsize = (10,5))
        sns.countplot(data = df, x = col, hue = 'Converted')
        plt.title(f'{col} vs Conversion')
        plt.xticks(rotation = 45)
        plt.tight_layout()
        plt.show()
```

```python
#
plot_cats(cat_cols)
```

```python
# Numeric feature correlation
numeric_df = df.select_dtypes(include=['number'])
plt.figure(figsize = (10,6))
sns.heatmap(numeric_df.corr(), annot = True, cmap = 'coolwarm', fmt = '.2f')
plt.title('Feature Correlation Heatmap')
plt.show()
```
